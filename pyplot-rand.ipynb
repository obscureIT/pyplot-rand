```python
#importing numpy

import numpy as np

#importing pyplot module of matplotlib package/library
import matplotlib.pyplot as plt 

# for determining the size of the plot view/display
from matplotlib.pyplot import figure
figure(figsize=(10, 8), dpi=100) 


N = 256
x = np.random.rand(N) #creating random numbers with numpy
y = np.random.rand(N)

# print(x)
# print(y)

colors = np.random.rand(N)
area = np.pi * (30 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=.80)
plt.show()
```
